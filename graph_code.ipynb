```python
import numpy as np# numpy is used to solve mathematical things such as matrix
import matplotlib.pyplot as plt# matplotlib is used in graph making or plotting
#here we give a new name to lib name
```

get the data/generate the data

```python
#both list value number shoulb be same
rainfall=[17,10,10,10,26,59]
months=['Jan','Feb','Mar','April','May','June']
```

visualize the data

```python
plt.style.available
```

```python
#theme
plt.style.use('ggplot')
```

```python
#plt has many methods, we can see by using plt.
plt.bar(months,rainfall,color='#ff77b4')# color optional h
```

```python
plt.pie(rainfall,labels=months,autopct="%1.1f%%",wedgeprops={'width':.4,'edgecolor':'#6666ff','linewidth':2},)
plt.show()
plt.legend()

plt.pie(rainfall,labels=months,startangle=90)
plt.legend(title="Months:")
plt.show()
```

```python
plt.boxplot(rainfall,vert=True,showmeans=True,meanline=True,meanprops={'color':'red'})
plt.show()
```

```python

#using matplotlib site
fig, ax = plt.subplots(figsize=(10,5))  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3],label='blue')  # Plot some data on the axes
ax.plot([1, 2, 3, 4], [2, 4, 5, 7],label='orange') 
ax.grid()
ax.legend()
```

```python

t=np.arange(-10,10,.2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(t,y,label='sinwave')
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t=np.arange(-10,10,.2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(t,y,label='sinwave',color='green',linestyle='--',linewidth='3',marker='o',markersize='4',
        markerfacecolor='red',alpha=.5)#alpha value between .2 to .5
t=np.arange(-10,10,.2)
y2=np.cos(t)
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(t,y2,label='coswave',color='green',linestyle='--',linewidth='3',marker='o',markersize='4',
        markerfacecolor='red',alpha=.5)#alpha value between .2 to .5
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')#MARKER=o,s,x,*,1

ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')#MARKER=o,s,x,*,1

```
